```python
data = pd.read_csv("social_network_ads.csv")
```

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
cov_matrix = data.cov()
```

```python
import seaborn as sns
```

```python
hisplot =sns.histplot(data['Age'], bins=30, kde=True, color="blue",alpha = 0.3)
for i in hisplot.containers:
    hisplot.bar_label(i,)
plt.show()    
```

```python
histplot1 =sns.histplot(data['EstimatedSalary'], bins=30, kde=True, color='red',alpha = 0.3)
for i in histplot1.containers:
    histplot1.bar_label(i,)
plt.show()    
```
